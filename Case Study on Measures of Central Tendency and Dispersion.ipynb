# 1. Find out how many males and females participated in the test.

```python
plt.figure(figsize=(8,8))
plt.title('Gender count in the test')
fig=sns.countplot(x='gender',data=data)
#to get the count displayed on the chart
for p in fig.patches:
    fig.annotate(p.get_height() , (p.get_x()+0.35, p.get_height()-250))
plt.show()
```

Insights: <br>
    518 females and 482 males participated in the test. 

# 2. What do you think about the students' parental level of education?

```python
plt.figure(figsize=(16,18))
plt.subplot(4,1,1)
plt.title('Distribution of Parental level of education w.r.to Student gender')
fig=sns.countplot(x='parental level of education',hue='gender',data=data)
plt.ylim(0,130)
#to get the count displayed on the chart
for p in fig.patches:
    fig.annotate(p.get_height() , (p.get_x()+0.15, p.get_height()+0.9))
plt.show()


plt.figure(figsize=(16,18))
plt.subplot(4,1,2)
plt.title('Distribution of Parental level of education w.r.to Race/ethnicity')
plt.ylim(0,90)
fig=sns.countplot(x='parental level of education',hue='race/ethnicity',data=data)
#to get the count displayed on the chart
for p in fig.patches:
    fig.annotate(p.get_height() , (p.get_x()+0.05, p.get_height()+0.9))
plt.show()

plt.figure(figsize=(16,18))
plt.subplot(4,1,3)
plt.title('Distribution of Parental level of education w.r.to Test preparation course')
plt.ylim(0,160)
fig=sns.countplot(x='parental level of education',hue='test preparation course',data=data)
#to get the count displayed on the chart
for p in fig.patches:
    fig.annotate(p.get_height() , (p.get_x()+0.13, p.get_height()+0.9))
plt.show()

plt.figure(figsize=(16,18))
plt.subplot(4,1,4)
plt.title('Distribution of Parental level of education w.r.to Lunch')
plt.ylim(0,160)
fig=sns.countplot(x='parental level of education',hue='lunch',data=data)
#to get the count displayed on the chart
for p in fig.patches:
    fig.annotate(p.get_height() , (p.get_x()+0.15, p.get_height()+0.75))
plt.show()
```

Insights:<br>1. Around 65% of the students have an “associate’s degree, high school or some college” as the parental level of education.<br>2. Among students with the same parental level of education, there are more female students in every category other than high school.<br>3. Students hailing from 'group A' race fare the worse in terms of parental level of education. They have only 3 parents with a master's degree.<br>4. Among students with the same parental level of education, more than 30% completed the test preparation course in each category other than high school.<br>5. Among students with the same parental level of education, less than 40% opted for a free/reduced lunch in each category other than master's degree.


```python
plt.figure(figsize=(16,18))
plt.subplot(4,1,1)
sns.boxplot(x='parental level of education',y='math score',hue='gender',data=data)
plt.title('Scoring variation in math score based on parental level of education')
plt.show()

plt.figure(figsize=(16,18))
plt.subplot(4,1,2)
sns.boxplot(x='parental level of education',y='reading score',hue='gender',data=data)
plt.title('Scoring variation in reading score based on parental level of education')
plt.show()

plt.figure(figsize=(16,18))
plt.subplot(4,1,3)
sns.boxplot(x='parental level of education',y='writing score',hue='gender',data=data)
plt.title('Scoring variation in writing score based on parental level of education')
plt.show()

plt.figure(figsize=(16,18))
plt.subplot(4,1,4)
sns.boxplot(x='parental level of education',y='avg. score',hue='gender',data=data)
plt.title('Scoring variation in average score based on parental level of education')
plt.show()
```

Insights:<br>1. Male students with a parental level of education of master's degree have a higher median than others in maths.<br>2. However, no student with a parental level of education of master's degree have scored cent percent in maths.<br>3. Among students of same parental level of education, male students have a higher median and more consistent performance than females in maths.<br>4. In the case of reading score, females have performed better among students of same parental level of education other than master's degree.<br>5. In the case of writing score, females have performed better among students of same parental level of education.<br>6. Students with a parental level of education of master's degree have a more consistent performance. <br>7.  Among students with parental level of education of 'some high school', the whiskers are a bit longer, which implies there are students who are performing poorly, as well as students that are performing well. 

```python
plt.figure(figsize=(16,18))
plt.subplot(4,1,1)
plt.title('Parental level of education v/s average score')
fig=sns.barplot(x='parental level of education',y='avg. score',hue='gender',ci=None,data=data)
plt.ylim(0,110)
#to get the count displayed on the chart
for p in fig.patches:
    fig.annotate(round(p.get_height(),2) , (p.get_x()+0.15, p.get_height()+0.75))
plt.show()

plt.figure(figsize=(16,18))
plt.subplot(4,1,2)
plt.title('Parental level of education v/s math score')
fig=sns.barplot(x='parental level of education',y='math score',hue='gender',ci=None,data=data)
plt.ylim(0,110)
#to get the count displayed on the chart
for p in fig.patches:
    fig.annotate(round(p.get_height(),2) , (p.get_x()+0.15, p.get_height()+0.75))
plt.show()

plt.figure(figsize=(16,18))
plt.subplot(4,1,3)
plt.title('Parental level of education v/s reading score')
fig=sns.barplot(x='parental level of education',y='reading score',hue='gender',ci=None,data=data)
plt.ylim(0,110)
#to get the count displayed on the chart
for p in fig.patches:
    fig.annotate(round(p.get_height(),2) , (p.get_x()+0.15, p.get_height()+0.75))
plt.show()

plt.figure(figsize=(16,18))
plt.subplot(4,1,4)
plt.title('Parental level of education v/s writing score')
fig=sns.barplot(x='parental level of education',y='writing score',hue='gender',ci=None,data=data)
plt.ylim(0,110)
#to get the count displayed on the chart
for p in fig.patches:
    fig.annotate(round(p.get_height(),2) , (p.get_x()+0.15, p.get_height()+0.75))
plt.show()
```

Insights:<br>1. Among students with the same parental level of education, female students have higher average reading and writing scores than their male counterparts;whereas male students have higher average math scores.
